# Digit Recognition using Logistic Regression

## Importing Dependencies 

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.datasets import load_digits
```

```python
#Loading Data

digits = load_digits()
```

```python
#Seeing data information

dir(digits)
```

```python
#printing first image's data in the form of 2-D Array

digits.data[0]
```

```python
#Printing the First Image of Data

plt.gray()
plt.matshow(digits.images[0])
```

```python
#Printing a Range of Images

for i in range(10) :
    plt.matshow(digits.images[i])
```

```python
#Printing a Range of Images

for i in range(5) :
    plt.matshow(digits.images[i])
```

```python
#Printing the Digits of First 5 Images

digits.target[0:5]
```

## Splitting the Data 

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.2)
```

```python
#Printing the Length of X_train

len(X_train)
```

```python
#Printing the Length of X_test

len(X_test)
```

## Importing Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

## Fitting the Model

```python
model.fit(X_train, y_train)
```

## Printing the Accuracy of Model

```python
print('Accuracy :', (model.score(X_test, y_test))*100)
```

## Now Predicting Some Data

```python
#Printing a Random Image

plt.matshow(digits.images[150])
```

```python
#Printing the Digit of the Above Image

digits.target[150]
```

```python
#Now Predicting the value of the Above Image

model.predict(digits.data[[150]])
```

```python
#Now Predicting the First Five Values of Data

model.predict(digits.data[0:5])
```

## Now Printing the Confusion Matrix

```python
pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(pred, y_test)
```

```python
cm
```

## Making a HeatMap of Confusion Matrix

```python
plt.figure(figsize=(10, 7))
sns.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
plt.show()
```
